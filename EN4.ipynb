## Question II

```python
#Calculer la fonction p(u|X) ;
def MCMC(u0, s, epoch):
    u = u0
    res = []
    for i in range(epoch):
        norm = sps.norm(u, s*s);
        u_s = norm.rvs(1)[0];
        p_0 = math.exp(-( u*u/8  + n*(X_mean - u)*(X_mean - u)/2))
        p_s = math.exp(-( u_s*u_s/8  + n*(X_mean - u_s)*(X_mean - u_s)/2))
        if p_s / p_0 >= 1:
            u = u_s;
        else:
            bern = sps.bernoulli(p_s / p_0);
            b = bern.rvs(1)[0];
            if b == 1:
                u = u_s
        res.append(u)
        print("u at ", i+1 , " epoch: ", u);
    return res;
```

```python
# Test la fonction 
u_visited =  MCMC( 2, 0.1, 10000)
```

```python
# Visualization

fig1 = figure(title='u Visited', width=490, height=300, background_fill_color="#fafafa")
hist, edges = np.histogram(u_visited, bins=20)
fig1.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
          fill_color="navy", line_color="white", alpha=0.5, legend="simulation")
fig1.yaxis.axis_label = 'count'
fig1.xaxis.axis_label = 'u'
fig1.grid.grid_line_color="white"
show(fig1)
```

## Question III

```python
def taux_de_deplacement_in_MCMC(u0, s, epoch):
    u = u0
    res = []
    deplacement_effectue_count = 0;
    deplacement_propose_count = 0
    for i in range(epoch):
        norm = sps.norm(u, s*s);
        u_s = norm.rvs(1)[0];
        p_0 = math.exp(-( u*u/8  + n*(X_mean - u)*(X_mean - u)/2))
        p_s = math.exp(-( u_s*u_s/8  + n*(X_mean - u_s)*(X_mean - u_s)/2))
        if p_s / p_0 >= 1:
            u = u_s;
        else:
            bern = sps.bernoulli(p_s / p_0);
            b = bern.rvs(1)[0];
            if b == 1:
                u = u_s
                deplacement_effectue_count += 1;
            deplacement_propose_count += 1
        res.append(u)
#         print("u at ", i+1 , " epoch: ", u);
    return deplacement_effectue_count / deplacement_propose_count
```

```python
print(taux_de_deplacement_in_MCMC(2.5, 0.1, 10000))
```

```python
cands = [0.1, 0.2, 0.5 ,1.0, 1.5, 2.0, 2.5]
winner = [];
for cand in cands:
    t = taux_de_deplacement_in_MCMC(2.3, cand, 10000) 
    if t < 0.25:
        winner.append(cand);
    print( "Le taux de acception de ", cand, " is ", t);
print("Les valeurs de s telles que le taux d'acception est inferieur Ã  25% est: ", winner);
```
